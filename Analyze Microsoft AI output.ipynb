```python
detection_cat = {'1': 'animal', '2': 'person', '3': 'vehicle'}

def parseDetections(det, cat):
    try:
        if len(det) == 0:
            return 0
        conf = []
        for d in det:
            if 'conf' in d.keys():
                if d['category'] == cat:
#                 conf+=str(d['conf'])+','
                    return 1
        return 0
    except:
        print(det)
        return 0
    
def parseEmpties(det):
    try:
        if len(det) > 0:
            return 0
        else:
            return 1
    except:
        return 0

#person, animal, vehicle
df['person'] = df['detections'].apply(lambda x: parseDetections(x, '2'))
df['animal'] = df['detections'].apply(lambda x: parseDetections(x, '1'))
df['vehicle'] = df['detections'].apply(lambda x: parseDetections(x, '3'))

df['empty'] = df['detections'].apply(lambda x: parseEmpties(x))
```

```python
def getMaxConf(det, cat):
#     print(cat)
    try:
        if len(det) == 0:
            return 0
        maxConf = 0
        for d in det:
            if 'conf' in d.keys():
                if d['category'] == cat:
                    if d['conf'] > maxConf:
                        maxConf = d['conf']
        return maxConf
    except:
        print(det)
        return 0
    
df['person_maxConf'] = df['detections'].apply(lambda x: getMaxConf(x, '2'))
df['animal_maxConf'] = df['detections'].apply(lambda x: getMaxConf(x, '1'))
df['vehicle_maxConf'] = df['detections'].apply(lambda x: getMaxConf(x, '3'))
```
