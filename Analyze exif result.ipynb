```python
file = open('results_exif12.out', 'r')

results = []

count = 0

while True:
    line1 = file.readline()
    if not line1:
        break
    line1 = list(reader([line1], delimiter=',', quotechar='"'))[0]

    line2 = file.readline()
    if not line2:
        break
#         print(line2)
    line2 = list(reader([line2.replace('\0', '')], delimiter=',', quotechar='"'))[0]
    
#     if len(line1) == 45:
#         print(line1)
#         print(line2)
#         input('Press Enter ...')

    hold = {line1[i]: line2[i] for i in range(len(line1))}

    results.append(hold)

    count += 1
    if count % 50000 == 0:
        print(count)
        
file.close()
```

# How to create path and trasfer?

```python
import os
from shutil import copyfile
```

```python
machine_path_to_mnt_parent_dl = '/Volumes/wildlifeimages/wildlifemontoringprogram'
machine_path_to_mnt_nonhuman_dl = '/Volumes/nonhumanimages'

for root in df_ai_w_metadta_notempty_nohuman_animallikely[50000:50010]['file'].to_list():
    # iterated over the path
    if not os.path.isfile(os.path.join(machine_path_to_mnt_nonhuman_dl, root)):
        break_path = root.split('/')
    #     print(break_path[:-1])
        for i, child in enumerate(break_path[:-1]):
            if os.path.isdir(os.path.join(machine_path_to_mnt_nonhuman_dl, '/'.join(break_path[:i+1]))):
                continue
            else:
                print('create path: ', os.path.join(machine_path_to_mnt_nonhuman_dl, '/'.join(break_path[:i+1])))
                os.mkdir(os.path.join(machine_path_to_mnt_nonhuman_dl, '/'.join(break_path[:i+1])))
        #tranfer the file
        src = os.path.join(machine_path_to_mnt_parent_dl, '/'.join(break_path))
        dst = os.path.join(machine_path_to_mnt_nonhuman_dl, '/'.join(break_path))
    #     print(src)
    #     print(dst)
        copyfile(src, dst)
    else:
        print('exists: ', os.path.join(machine_path_to_mnt_nonhuman_dl, '/'.join(break_path)))
```
