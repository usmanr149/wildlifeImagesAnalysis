```python
import os
import glob
import pandas as pd
```

```python
# Site Name	SD    	   Subfolder   # photos

sitename_subfolder = []


for root, dirs, files in os.walk("/Volumes/nonhumanimages/", topdown=False):
#     for name in files:
#         print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
        sitename_subfolder.append(os.path.join(root, name))
```

```python
len(sitename_subfolder)
```

```python
sitename = []
sd = []
subfolder = []
file_numbers = []

for s in sitename_subfolder:
    if '/Backups' not in s:
        split_name = s.split('/')
        sitename.append(split_name[3])
        if len(split_name) > 4:
#             print(split_name)
            sd.append(split_name[4])
            if len(split_name) > 5:
                subfolder.append('/'.join(split_name[5:]))
            else:
                subfolder.append('')
        else:
            sd.append('')
            subfolder.append('')
        print(s)
        path = s + '/*.JPG'
        print(path)
        JPGs = glob.glob(path)
        file_numbers.append(len(JPGs))
```

```python
df = pd.DataFrame({'Site Name': sitename, 'SD': sd, 'Subfolder': subfolder, '# photos': file_numbers} )

df = df[df['# photos'] != 0]
```
